```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
```

```python
df = pd.read_csv('data/DB.csv')
df
```

```python

years = df.columns[1::].tolist()
countrys = df['Country'].tolist()
populations = []
for country in countrys:
    pop = (df[df['Country'] == country].iloc[:, 1:].values[0]/1e6).tolist()

    plt.plot(years, pop)
    plt.title(country)
    plt.xlabel('Years')
    plt.xticks(years[::5], rotation=45)
    plt.ylabel('Popultion (milions)')
    plt.grid(True)
    plt.show()
```
